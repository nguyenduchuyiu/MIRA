```python
import numpy as np

# Load data from noise.txt
noise_data = np.loadtxt('noise.txt')

# Load data from sound.txt
sound_data = np.loadtxt('sound.txt')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Plot distribution of noise
plt.subplot(1, 2, 1)
sns.histplot(noise_data, bins=30, kde=True, color='blue', stat='density')
plt.title('Distribution of Noise')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot distribution of sound
plt.subplot(1, 2, 2)
sns.histplot(sound_data, bins=30, kde=True, color='orange', stat='density')
plt.title('Distribution of Sound')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plots
plt.tight_layout()
plt.show()
```
